```python
import pandas as pd
```

```python

def clean_data(file_path):
    # Load the dataset
    try:
        df = pd.read_csv(file_path)
        df=pd.DataFrame(df)
        print(df.isnull().sum())
        if df.isnull().sum().sum()>0:
            print("Null values found. Dropping rows with null values.")
            df.dropna(inplace=True)
            
        print("duplicate:",df.duplicated().sum())
        if df.duplicated().sum()>0:
            df.drop_duplicates(inplace=True)
            
    except Exception as e:
        print(f"Error loading data: {e}")
        return None
    return df
```

```python
def passed(score):
    try:
        if score >= 50 and score <= 100:
            return "true"
        else:
            return 'False'
        
    except Exception as e:
        print(f"Error processing score: {e}")
            
    
```

```python
def assigned_column(df):
    try:
        df["passed"] = df['score'].apply(passed)
        return df
    except Exception as e:
        print(f"Error assigning pass/fail: {e}")
```

```python
def score_catagory(score):
    try:
        if score >= 80 and score <=100:
            return "high"
        elif score >= 50 and score < 80:
            return "mid"
        else:
            return "low"
    except Exception as e:
        print(f"Error categorizing score: {e}")
```

```python
def assign_score_category(df):
    try:
        df["score_category"] = df['score'].apply(score_catagory)
        return df
    except Exception as e:
        print(f"Error assigning score category: {e}")
```

```python
def applying_rank(df):
    try:
        df['ranked'] = df['score'].rank(ascending=False)
        return df
    except Exception as e:
        
        print(f"Error ranking scores: {e}")
```

```python
def groupbygrade(df):
    try:
        grade_summary = df.groupby('grade')['score'].agg(['mean', 'min', 'max', 'count'])
        return grade_summary
    except Exception as e:
        print(f"Error grouping by grade: {e}")
    
```

```python
def main():
    try:
        file_path = '../task1/cleaned_students.csv'  # Replace with your actual file path
        cleaned_data = clean_data(file_path)
        if cleaned_data is not None:
            df = assigned_column(cleaned_data)
            df=assign_score_category(df)
            df=groupbygrade(df)
            print(df)
            
            df=df.sort_values(by='rank').reset_index(drop=True)
            df=applying_rank(df)
            df.to_csv('final_students.csv', index=True)
            
            
            print("top 5:",df.head())
            # Display the first few rows of the final DataFrame
    except Exception as e:
        print(f"Error in main execution: {e}")
    
if __name__ == "__main__":
    main()
```
